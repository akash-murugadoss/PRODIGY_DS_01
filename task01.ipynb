```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/World Population (2022) .csv")

gender_counts = df['Country'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.show()

plt.hist(df['Med.Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Med.Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```
